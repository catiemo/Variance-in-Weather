```python
import codecademylib3_seaborn
import pandas as pd
import numpy as np
from weather_data import london_data

# Loading London DataFrame
print(london_data.head())

# Viewing rows 100 through 199 of London DataFrame 
print(london_data.iloc[100:200])

# Checking number of data points
print(len(london_data))

# Selecting the Temperature Column
temp = london_data.TemperatureC

# Calculating Average Temperature
average_temp = np.mean(temp)
print(average_temp)

# Calculating Variance of Temperature
temperature_var = np.var(temp)
print(temperature_var)

# Calculating Standard Deviation of Temperature
temperature_standard_deviation = np.std(temp)
print(temperature_standard_deviation)

# Filtering Temperature DataFrame by month(June)
june = london_data.loc[london_data.month == 6]['TemperatureC']

# Finding average of temperatures in June
print(np.mean(june))

# Finding standard deviation of temperatures in June
print(np.std(june))

# Filtering Temperature DataFrame by month(July)
july = london_data.loc[london_data.month == 7]['TemperatureC']

# Finding average of temperatures in July
print(np.mean(july))

# Finding standard deviation of temperatures in July
print(np.std(july))

# Finding the average, standard deviation, variance of temperatures of each month
for i in range(1, 13):
    month = london_data.loc[london_data.month == i]['TemperatureC']
    print('The mean temperature in month ' + str(i) + ' is ' + str(np.mean(month)))
    print('The standard deviation temperature in month ' + str(i) + ' is ' + str(np.std(month)) + '\n')

# Finding the rainest month
rainest_month = london_data.groupby('month').Humidity.mean().reset_index()
print(rainest_month[rainest_month.Humidity == rainest_month.Humidity.max()])
```
